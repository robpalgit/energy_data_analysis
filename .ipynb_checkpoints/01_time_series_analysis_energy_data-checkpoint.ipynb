```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sb
plt.style.use('bmh')

from pandas.plotting import lag_plot, scatter_matrix, autocorrelation_plot
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

## Load dataset

```python
df = pd.read_csv('base_para_prediccion.csv', parse_dates=['fecha'])
df
```

```python
df.info()
```

```python
df.nunique()
```

## Some data cleaning...

```python
# Delete rows where hora == 25
to_delete = df[df['hora']==25].index.tolist()
df.drop(to_delete, axis=0, inplace=True)
print('Number of rows deleted: {}'.format(len(to_delete)))
print(df.shape[0])
```

```python
# Check for duplicates
subset = [
    'nemotecnico_se',
    'fecha',
    'hora'
]

bool_mask_01 = df.duplicated(subset=subset)
print('Number of duplicate rows: ', df[bool_mask_01].shape[0])
```

## Create new variables

```python
# Change hour "24" to hour "0"
df['hora'] = df['hora'].apply(lambda x: 0 if x==24 else x)

df['fecha_hora'] = df['fecha'] + pd.to_timedelta(df['hora'], unit='H')
```

```python
df['año'] = df['fecha'].dt.year
df['mes'] = df['fecha'].dt.month
# Dia del mes
df['dia'] = df['fecha'].dt.day
# Semana del año (1 - 52)
df['semana_año'] = df['fecha'].dt.isocalendar().week
# Dia calendario (1 - 365)
df['dia_calendario'] = df['fecha'].dt.dayofyear
# Dia de la semana (0 - 6)
df['dia_semana'] = df['fecha'].dt.weekday
# Fin de semana vs. dia de semana (Weekend vs Business)
df['fin_semana_si_no'] = df['dia_semana'].map(lambda x: 1 if x in [5,6] else 0)
```

### 1. Multiple line plots

```python
def line_plots_by_month(se_code, num_var):
    months_dict = {
        1: 'January',
        2: 'February',
        3: 'March',
        4: 'April',
        5: 'May',
        6: 'June',
        7: 'July',
        8: 'August',
        9: 'September',
        10: 'October',
        11: 'November',
        12: 'December'
    }
    data = df[df['nemotecnico_se']==se_code][['fecha_hora', num_var]]
    series = data.sort_values(by=['fecha_hora']).set_index('fecha_hora').squeeze()
    
    groups = series.groupby(pd.Grouper(freq='M'))
    for name,group in groups:
        pd.Series(group.values).plot(figsize=(12,2))
        plt.title('{} - {}'.format(num_var, months_dict[name.month]))
        plt.show()
```

```python
line_plots_by_month('SE004T002', 'cmg_real')
```

```python
def line_plots_by_month(se_code, num_var):
    months_dict = {
        1: 'January',
        2: 'February',
        3: 'March',
        4: 'April',
        5: 'May',
        6: 'June',
        7: 'July',
        8: 'August',
        9: 'September',
        10: 'October',
        11: 'November',
        12: 'December'
    }
    variables = ['mes', 'fecha_hora', num_var]
    data = df[df['nemotecnico_se']==se_code][variables]
    months = data['mes'].unique()
    for month in months:
        month_data = data[data['mes']==month][variables[1:]]
        month_data.sort_values(by='fecha_hora')\
        .set_index('fecha_hora')\
        .plot(figsize=(12,2), legend=False)
        plt.title('{} - {}'.format(num_var, months_dict[month]))
        plt.xlabel("")
        plt.xticks(ticks=[])
        plt.show()
```

```python
line_plots_by_month('SE004T002', 'cmg_real')
```

```python
def line_plots_by_month(se_code, num_var):
    months_list = [
        'January',
        'February',
        'March',
        'April',
        'May',
        'June'
    ]
    data = df[df['nemotecnico_se']==se_code][['fecha_hora', num_var]]
    series = data.sort_values(by=['fecha_hora']).set_index('fecha_hora').squeeze()
    
    groups = series.groupby(pd.Grouper(freq='M'))
    months = pd.concat([pd.DataFrame(x[1].values) for x in groups], axis=1)
    months = pd.DataFrame(months)
    months.columns = months_list
    
    months.plot(subplots=True, legend=True, figsize=(12,10))
    plt.show()
```

```python
line_plots_by_month('SE004T002', 'cmg_real')
```

```python
def multiple_boxplots(num_var, cat_var, se_code=None, rotate_xticks=False):
    if se_code is not None:
        data = df[df['nemotecnico_se']==se_code]
    else:
        data = df
    
    plt.figure(figsize=(10,5))
    sb.boxplot(x=cat_var, y=num_var, data=data)
    if rotate_xticks:
        plt.xticks(rotation=45, ha='right')
    plt.show()
```

### 3. Multiple histograms

```python
def multiple_histograms(num_var):
    fig, axes = plt.subplots(10, 3, figsize=(10, 20))
    se_code_list = df['nemotecnico_se'].unique().tolist()

    for ax, se_code in zip(axes.flatten(), se_code_list):
        series = df[df['nemotecnico_se']==se_code][num_var]
        sb.histplot(x=series, kde=True, ax=ax)
        ax.set_xlabel(se_code)
    
    fig.tight_layout()
```

```python
multiple_histograms('cmg_real')
```

### 4. Time Series Heat Maps

```python
def time_series_heatmap(se_code, num_var, month:int):
    months_list = [
        'January',
        'February',
        'March',
        'April',
        'May',
        'June'
    ]
    data = df[
        (df['nemotecnico_se']==se_code) &
        (df['mes']==month)
    ][['fecha_hora', num_var]].sort_values(by='fecha_hora').set_index('fecha_hora').squeeze()
    
    groups = data.groupby(pd.Grouper(freq='D'))
    days = pd.DataFrame()

    for name,group in groups:
        days[name.day] = group.values
    
    fig = plt.figure(figsize=(7,6))
    ax = fig.add_subplot(1, 1, 1)
    i = ax.imshow(days, interpolation=None, aspect='auto', cmap='inferno')
    ax.set_yticks(ticks=np.arange(0,24))
    fig.colorbar(i)
    plt.show()
```

```python
time_series_heatmap('SE004T002', 'cmg_real', 1)
```

```python
def time_series_heatmap(se_code, num_var):
    
    data = df[
        (df['nemotecnico_se']==se_code)
    ][['fecha_hora', 'dia_calendario', num_var]].sort_values(by='fecha_hora')
    
    groups = data[['dia_calendario', num_var]].groupby(pd.Grouper(key='dia_calendario'))
    days = pd.DataFrame()

    for name,group in groups:
        days[name] = group[num_var].values
    
    fig = plt.figure(figsize=(18,4))
    ax = fig.add_subplot(1, 1, 1)
    i = ax.imshow(days, interpolation=None, aspect='auto', cmap='inferno')
    ax.set_title('Time series heatmap for {} - Substation {}'.format(num_var, se_code))
    ax.set_xlabel('Calendar days')
    ax.set_ylabel('Hours')
    ax.set_xticks(ticks=np.arange(0, 181, 30))
    ax.set_yticks(ticks=np.arange(0, 24, 6))
    fig.colorbar(i)
    plt.show()
```

```python
time_series_heatmap('SE004T002', 'cmg_real')
```

```python
for sub in df['nemotecnico_se'].unique():
    time_series_heatmap(sub, 'cmg_real')
```

### 5. Time Series Lag Scatter Plots

```python
def custom_lag_plot(num_var, se_code, lag=1):
    data = df[
        (df['nemotecnico_se']==se_code)
    ][['fecha_hora', num_var]].sort_values(by='fecha_hora').set_index('fecha_hora').squeeze()
    
    lag_plot(data, lag=lag)
    plt.show()
```

```python
custom_lag_plot('cmg_real', 'SE004T002')
```

```python
custom_lag_plot('cmg_real', 'SE004T002', lag=4)
```

```python
custom_lag_plot('cmg_real', 'SE004T002', lag=7)
```

### 6. Time Series Autocorrelation Plots

```python
def custom_autocorrelation_plot(num_var, se_code):
    data = df[
        (df['nemotecnico_se']==se_code)
    ][['fecha_hora', num_var]].sort_values(by='fecha_hora').set_index('fecha_hora').squeeze()
    
    fig = plt.figure(figsize=(18,7))
    ax = fig.add_subplot(1, 1, 1)
    autocorrelation_plot(data)
    ax.set_xticks(ticks=np.arange(0, 4500, 100))
    plt.show()
```

```python
custom_autocorrelation_plot('cmg_real', 'SE004T002')
```
